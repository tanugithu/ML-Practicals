```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
data = load_iris()
x = data.data
y = data.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
explained_varience = pca.explained_variance_ratio_
print(f'Explained varience by each component: {explained_varience}')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y, cmap='viridis', edgecolors= 'k',s=100)
plt.title('PCA of Iris Dataset')
plt.xlabel(f'principal Component 1({explained_varience[0]:.2f} varience)')
plt.ylabel(f'principal Component 2({explained_varience[0]:.2f} varience)')
plt.colorbar()
plt.grid(True)
plt.show()
```
